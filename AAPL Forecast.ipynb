```python
import yfinance as yf
```

```python
apple = yf.Ticker("AAPL")

apple = apple.history(period="5y")
```

```python
#we predict using the previous day close price

import pandas as pd

apple["Next Day Close"] = apple["Close"].shift(-1)
```

```python
apple = apple.drop(["Dividends","Stock Splits"],axis =1)
```

```python
apple.head()
```

```python
apple=apple.dropna()
```

```python
train = apple.iloc[:-100]
test = apple.iloc[-100:]
```

```python
#Splitting the features into independent and target features

x_train=train [["Open","High","Low","Close","Volume"]]
y_train=train["Next Day Close"]
x_test=test [["Open","High","Low","Close","Volume"]]
y_test=test["Next Day Close"]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
